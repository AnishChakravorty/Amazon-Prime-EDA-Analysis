### Outlier Detection 

```python
# Identify numerical columns
numerical_columns = ['imdb_score']

# Find outliers using IQR method and display bounds
outlier_bounds = {}
for column in numerical_columns:
    Q1 = df_exploded[column].quantile(0.25)
    Q3 = df_exploded[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outlier_bounds[column] = (lower_bound, upper_bound)

# Display lower and upper bounds for each column
for column, bounds in outlier_bounds.items():
    print(f"{column} - Lower Bound: {bounds[0]}, Upper Bound: {bounds[1]}")
```

```python
# Identify numerical columns
numerical_columns = ['tmdb_score']

# Find outliers using IQR method and display bounds
outlier_bounds = {}
for column in numerical_columns:
    Q1 = df_exploded[column].quantile(0.25)
    Q3 = df_exploded[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outlier_bounds[column] = (lower_bound, upper_bound)

# Display lower and upper bounds for each column
for column, bounds in outlier_bounds.items():
    print(f"{column} - Lower Bound: {bounds[0]}, Upper Bound: {bounds[1]}")
```

Outliers should not be removed in this case because they represent extreme values that could provide valuable insights into the dataset. For example, in the context of movies and TV shows, outliers might indicate exceptionally popular or highly rated content, which could be critical for understanding audience preferences or identifying standout titles. Removing these data points could lead to a loss of important information and skew the analysis.

Additionally, the threshold of 10 is significant because it ensures that we retain enough data points for meaningful analysis. If we were to remove outliers below or above this threshold, we might inadvertently exclude rare but important cases, such as niche genres or unique production countries, which could provide valuable insights into the diversity and reach of the content library.
